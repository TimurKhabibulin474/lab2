### Работа с NaN
Часто данные содержат пропущенные значения, которые представляются как NaN (Not a Number). Pandas предоставляет удобные методы для работы с ними.

Проверка на наличие NaN. Методы fillna и dropna возвращают новые Dataframe, проверьте нет ли в них NaN

```python
# Вывод строк содержащих NaN
data_titanic[data_titanic.isna().any(axis=1)]
```

Заполнение NaN

```python
# Заполнение NaN определенным значением  нулем
data_titanic_fillna = data_titanic.fillna(0)
data_titanic_fillna.head()
```

Удаление строк с NaN

```python
# Удаление строк, содержащих NaN
data_titanic_dropna = data_titanic.dropna()
data_titanic_dropna.head()
```

## Часть 3: Обработка DataFrame
### Выбор данных
Pandas позволяет выбирать данные по индексам, меткам столбцов и условиям.

```python
# Столбец Name
titanic_name = data_titanic["Name"]

# Столбцы Name и Survived
titanic_is_survived = data_titanic[['Name', 'Survived']]

# Первые 100 строк
titanic_subdata = data_titanic.loc[range(0, 100)]

# Строки с мужчинами старше 30
titanic_male_30 = data_titanic.query("Sex == 'male' & Age > 30.0")
```

### Сортировка данных
Сортировка данных по значениям столбцов.

```python
# Сортировка данных по столбцу 'Age' по возрастанию
titanic_sort_Age = data_titanic.sort_values("Age")
```

### Группировка данных
Pandas также позволяет группировать данные и выполнять агрегирующие операции.

```python
# Доля выживших среди всех PClass
data_titanic.groupby("Pclass") \
  .agg({"Survived": ["sum", "count"]}) \
  .apply(lambda x: x.Survived["sum"] / x.Survived["count"], axis=1)
```

Выберите только те строки, где значение в столбце 'Age' больше 30.
Отсортируйте данные по столбцу 'Fare' в порядке убывания.
Сгруппируйте данные по столбцу 'Pclass' и вычислите средний возраст ('Age') для каждого класса.
